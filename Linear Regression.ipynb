```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv('heightweight.csv')
```

```python
x=df.loc[:,'height']
y=df.loc[:,'weight']
```

```python
m,c,r,p,std_err=stats.linregress(x,y)
```

```python
def y_value(x):
    return m*x+c
model=list(map(y_value,x))
plt.scatter(x,y,c='g',label='Datas')
plt.plot(x,model,c='r',label='Regression')
predicted=float(input("Enter Height:"))
print("predicted Weight:",y_value(predicted))
plt.scatter(predicted,y_value(predicted),marker='o',color="blue",label="Predicted Value")
plt.legend(bbox_to_anchor=(1.4,1.03))
plt.show()
```
